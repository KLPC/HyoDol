## 변수 치환
1. `sex` : 여:1, 남0
2. `type` : 기초생활수급 : 0 , 차상위 :1
3. `pa1~pa10` , `sgd`가 들어간, `A1~A12`, `p_a`가 들어간 : 예:1, 아니오 : 0
4. `dq1~dq2` : 유:1, 무:0
5. `dq3` : 무자녀:0
6. `dq4` : 상:2, 중:1, 하:0
7. `D1,D2,D5,D6,D7`, `p_d`가 들어간, `P_D7` : 일정 하지 않음:1,때에 따라:2,매일 규칙적:3
8. `D3` : 
    - 매일 정확한 시간에 약을 복용하고 있음 : 3
    - 가끔 약 먹는 것을 잊어 먹을 때가 있음 : 2
    - 매번 약 먹는 것을 잊어 먹음 : 1

9. `D4` :
    - 매일 규칙적 세끼 식사하고 있음 : 3
    - 가끔 식사시간 놓침 : 2
    - 거의 매번 식사 시간 놓침 :1

10.`D8` :
    - 매우 긍정적 : 3
    - 조금 있음 : 2
    - 거의 없음 : 1


11. `j1~j4` :
    - 그렇다 : 3
    - 보통  : 2
    - 그렇지 않다 : 1


## 변수 생성
12. 변수 생성
- `type3` $\rightarrow$ `type3_binary` : '없음'이면 0, 나머지는 1

## 변수 인코딩
13. 라벨인코딩할 칼럼
- `['type2','type3','house','dq5','dq6']`
- `'dq8'` 포함한 칼럼

## 변수 제거
14. unique한 칼럼 제거
15. `i1~i11`제거
16. `DISE1~DISE4` 제거
17. `born`(탄생년도) 제거 ( 어차피 나이(`age`) 있음 )

```python
def cols_with_str(df,str_list):
    col_list=[]
    cols = df.columns
    for str_ in str_list:
        temp=list(cols[cols.str.contains(str_)])
        col_list.extend(temp)
    return col_list
```

### 변수 치환

```python
survey
```

```python
# sex
survey['sex']=survey['sex'].replace({'여':1,'남':0})

# type1
survey['type1']=survey['type1'].replace({'기초생활수급':0,'차상위':1})

# pa1가 들어간 / sgd가 들어간 / A1~A12 /p_a들어간
cols_=cols_with_str(survey,['pa','sgd','p_a'])+['A'+str(i) for i in range(1,13)]
survey[cols_]=survey[cols_].replace({'예':1,'아니오':0})

# dq1~dq2
survey[['dq1','dq2']]=survey[['dq1','dq2']].replace({'유':1,'무':0})

# dq3
survey['dq3']=survey['dq3'].replace({'무자녀':0})

# dq4
survey['dq4']=survey['dq4'].replace({'상':2,'중':1,'하':0})

# dq7_1
# ( fill NA = 0 할 것! )
survey['dq7_1']=survey['dq7_1'].replace({'아침 식전':1,'복용약 없음':0})
survey['dq7_2']=survey['dq7_2'].replace({'아침 식후':1,'복용약 없음':0})
survey['dq7_3']=survey['dq7_3'].replace({'점심 식전':1,'복용약 없음':0})
survey['dq7_4']=survey['dq7_4'].replace({'점심 식후':1,'복용약 없음':0})
survey['dq7_5']=survey['dq7_5'].replace({'저녁 식전':1,'복용약 없음':0})
survey['dq7_6']=survey['dq7_6'].replace({'저녁 식후':1,'복용약 없음':0})
survey['dq7_7']=survey['dq7_7'].replace({'취침 전':1,'복용약 없음':0})
survey[cols_with_str(survey,['dq7'])].fillna(0,inplace=True)

# D1,D2,D5,D6,D7, p_d가 들어간 / p_D7
cols_ = cols_with_str(survey,['p_d']) + ['D1','D2','D5','D6','D7','P_D7']
survey[cols_]=survey[cols_].replace({'일정 하지 않음':1,'때에 따라':2,'매일 규칙적':3})

# D3
survey['D3']=survey['D3'].replace({'매일 정확한 시간에 약을 복용하고 있음':3,
                                   '가끔 약 먹는 것을 잊어 먹을 때가 있음':2,
                                   '거의 매번 약 먹는 것을 잊어 먹음':1})

# D4
survey['D4']=survey['D4'].replace({'매일 규칙적 세끼 식사하고 있음':3,
                                  '가끔 식사시간 놓침':2,
                                  '거의 매번 식사 시간 놓침':1})
    

# D8
survey['D8']=survey['D8'].replace({'매우 긍정적':3,'조금 있음':2,'거의 없음':1})

# j1~j4
cols_ = ['j1','j2','j3','j4']
survey[cols_]=survey[cols_].replace({'그렇다':3,'보통':2,'그렇지 않다':1})

```

### 변수 제거

```python
survey[survey.columns[survey.columns.str.contains('dq7_')]]=survey[survey.columns[survey.columns.str.contains('dq7_')]].fillna(0)

drop_cols=[]
drop_cols.extend(list(survey.columns[survey.nunique()==1]))
drop_cols.extend(['i'+str(i) for i in range(1,11)])
drop_cols.extend(['DISE'+str(i) for i in range(1,4)])
drop_cols.extend(['born'])
survey=survey.drop(drop_cols,axis=1)
```

### 변수 인코딩

- dq8~로 시작하는, 질병 종류 묻는 문제임에도 불구하고 '숫자'가 들어간 이상한 경우는 NA로
- label인코딩 된 값 앞에 'cat'가 붙게 끔!

```python
from sklearn import preprocessing

def label_encode(survey,col):
    le = preprocessing.LabelEncoder()
    survey[col][pd.to_numeric(survey[col], errors='coerce').notnull()]=np.nan
    le.fit(survey[col])
    survey[col]=le.transform(survey[col])
    survey[col]=['cat'+str(i) for i in survey[col]]
```

```python
cols_for_label_encode=['GP','sub_gp','type2','type3','house','dq5','dq6']
cols_for_label_encode.extend(survey.columns[survey.columns.str.contains('dq8')])

survey['dq8_1']=survey['dq8_1'].replace({20:np.nan,44:np.nan,91:np.nan})
survey['dq8_2']=survey['dq8_2'].replace({4:np.nan,39:np.nan,49:np.nan,91:np.nan})
survey['dq8_3']=survey['dq8_3'].replace({44:np.nan})


for col in cols_for_label_encode:
    label_encode(survey,col)
```

```python
survey['reli']=['cat'+str(i) for i in  survey['reli'].fillna('999')]
survey['reli']=survey['reli'].replace({'cat999':np.nan})
```
